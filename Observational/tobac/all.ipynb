### _PERFORMING STATISTICS ON CELL_

```python
mean_max_dBZ_threshold_20 = []; std_max_dBZ_threshold_20 = []; rel_times_threshold_20 = [];
mean_mean_dBZ_threshold_20 = []; std_mean_dBZ_threshold_20 = []; rel_times1_threshold_20 = [];
mean_area_threshold_20 = []; std_area_threshold_20 = []; rel_times11_threshold_20 = [];

for indexes, time in enumerate(np.unique(first_landfall_threshold_20['relative_time'])):
    # Sorting based on same relative times
    sort_relative_times_threshold_20 = first_landfall_threshold_20[first_landfall_threshold_20['relative_time'] == time]
    # Find each relative time associated with a particular reflectivity value
    # print(sort_relative_times['relative_time'].values, sort_relative_times['max dBZ'].values)
    
    
    mean_threshold_20 = np.mean(sort_relative_times_threshold_20['max dBZ'].values)
    means_threshold_20 = np.mean(sort_relative_times_threshold_20['mean dBZ'].values)
    meanss_threshold_20 = np.nanmean(sort_relative_times_threshold_20['area($km^2$)'].values)
    
    std_threshold_20 = np.std(sort_relative_times_threshold_20['max dBZ'].values)
    stds_threshold_20 = np.std(sort_relative_times_threshold_20['mean dBZ'].values)
    stdss_threshold_20 = np.nanstd(sort_relative_times_threshold_20['area($km^2$)'].values)
    
    mean_max_dBZ_threshold_20.append(mean_threshold_20)
    mean_mean_dBZ_threshold_20.append(means_threshold_20)
    mean_area_threshold_20.append(meanss_threshold_20)
    
    std_max_dBZ_threshold_20.append(std_threshold_20)
    std_mean_dBZ_threshold_20.append(stds_threshold_20)
    std_area_threshold_20.append(stdss_threshold_20)
    
    rel_time_threshold_20 = np.mean(sort_relative_times_threshold_20['relative_time'].values)
    rel_times_threshold_20.append(rel_time_threshold_20)
    rel_times1_threshold_20.append(rel_time_threshold_20)
    rel_times11_threshold_20.append(rel_time_threshold_20)
    # print(np.mean(sort_relative_times['relative_time'].values)),print(mean)
```
