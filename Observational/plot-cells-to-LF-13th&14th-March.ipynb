```python
mean_max_dBZ_threshold_20 = []; std_max_dBZ_threshold_20 = []; rel_times_threshold_20 = [];
mean_mean_dBZ_threshold_20 = []; std_mean_dBZ_threshold_20 = []; rel_times1_threshold_20 = [];
mean_area_threshold_20 = []; std_area_threshold_20 = []; rel_times11_threshold_20 = [];

for indexes, time in enumerate(np.unique(first_landfall_threshold_20['relative_time'])):
    # Sorting based on same relative times
    sort_relative_times_threshold_20 = first_landfall_threshold_20[first_landfall_threshold_20['relative_time'] == time]
    # Find each relative time associated with a particular reflectivity value
    # print(sort_relative_times['relative_time'].values, sort_relative_times['max dBZ'].values)
    
    
    mean_threshold_20 = np.mean(sort_relative_times_threshold_20['max dBZ'].values)
    means_threshold_20 = np.mean(sort_relative_times_threshold_20['mean dBZ'].values)
    meanss_threshold_20 = np.nanmean(sort_relative_times_threshold_20['area($km^2$)'].values)
    
    std_threshold_20 = np.std(sort_relative_times_threshold_20['max dBZ'].values)
    stds_threshold_20 = np.std(sort_relative_times_threshold_20['mean dBZ'].values)
    stdss_threshold_20 = np.nanstd(sort_relative_times_threshold_20['area($km^2$)'].values)
    
    mean_max_dBZ_threshold_20.append(mean_threshold_20)
    mean_mean_dBZ_threshold_20.append(means_threshold_20)
    mean_area_threshold_20.append(meanss_threshold_20)
    
    std_max_dBZ_threshold_20.append(std_threshold_20)
    std_mean_dBZ_threshold_20.append(stds_threshold_20)
    std_area_threshold_20.append(stdss_threshold_20)
    
    rel_time_threshold_20 = np.mean(sort_relative_times_threshold_20['relative_time'].values)
    rel_times_threshold_20.append(rel_time_threshold_20)
    rel_times1_threshold_20.append(rel_time_threshold_20)
    rel_times11_threshold_20.append(rel_time_threshold_20)
    # print(np.mean(sort_relative_times['relative_time'].values)),print(mean)
```

## _CALCULATING N_CELLS AS A FUNCTION OF RELATIVE TIME_

```python
df5 = first_landfall_threshold_5.copy()
df10 = first_landfall_threshold_10.copy()
df15 = first_landfall_threshold_15.copy()
df20 = first_landfall_threshold_20.copy()

# sort dataframe by relative time -- ascending order
sorting_df5 = df5.sort_values(by=['relative_time'])
sorting_df10 = df10.sort_values(by=['relative_time'])
sorting_df15 = df15.sort_values(by=['relative_time'])
sorting_df20 = df20.sort_values(by=['relative_time'])

# get the relative time only
relative_time5 = sorting_df5['relative_time']
relative_time10 = sorting_df10['relative_time']
relative_time15 = sorting_df15['relative_time']
relative_time20 = sorting_df20['relative_time']

# get the number of counts based on the relative time
number_of_cell5 = relative_time5.value_counts()
relative_times5 = number_of_cell5.index # get relative time
number_cells5 = number_of_cell5.values # the number of cell based on relative time

number_of_cell10 = relative_time10.value_counts()
relative_times10 = number_of_cell10.index # get relative time
number_cells10 = number_of_cell10.values # the number of cell based on relative time

number_of_cell15 = relative_time15.value_counts()
relative_times15 = number_of_cell15.index # get relative time
number_cells15 = number_of_cell15.values # the number of cell based on relative time

number_of_cell20 = relative_time20.value_counts()
relative_times20 = number_of_cell20.index # get relative time
number_cells20 = number_of_cell20.values # the number of cell based on relative time

# len(relative_times), len(number_cells)
all_cells_count_landfall5 = pd.DataFrame(data={'relative time(min)': relative_times5, 
                                         'number cells': number_cells5})

all_cells_count_landfall10 = pd.DataFrame(data={'relative time(min)': relative_times10, 
                                         'number cells': number_cells10})

all_cells_count_landfall15 = pd.DataFrame(data={'relative time(min)': relative_times15, 
                                         'number cells': number_cells15})

all_cells_count_landfall20 = pd.DataFrame(data={'relative time(min)': relative_times20, 
                                         'number cells': number_cells20})
```
