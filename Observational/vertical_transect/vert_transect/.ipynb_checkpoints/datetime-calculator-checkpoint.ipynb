```python
def calculate_ct(start_date, end_date):
    # Convert start and end dates to datetime objects
    start_datetime = datetime.strptime(start_date, "%H:%M:%S %d %b %Y")
    end_datetime = datetime.strptime(end_date, "%H:%M:%S %d %b %Y")

    # Define the start and end years for the range
    start_year = 2019
    end_year = 2020

    # Define the start and end dates within the specified year range
    start_date_within_range = datetime(start_year, 12, 1)
    end_date_within_range = datetime(end_year, 5, 31)

    # Calculate the day values based on the adjusted start and end dates
    start_day = (start_datetime - start_date_within_range).days
    end_day = (end_datetime - start_date_within_range).days

    # Calculate the start and end minutes within a day
    start_minute = start_datetime.hour * 60 + start_datetime.minute
    end_minute = end_datetime.hour * 60 + end_datetime.minute

    # Adjust the start and end minutes if the seconds are greater than or equal to 30
    if start_datetime.second >= 30:
        start_minute += 1

    if end_datetime.second >= 30:
        end_minute += 1

    # Calculate the length by subtracting the start minute from the end minute
    length = end_minute - start_minute

    # Construct the output list with day values, start and end minutes, and length
    ct = [start_day, start_minute, end_day, end_minute, length]
    return ct


# Data containing the ID, start date, and end date
# kazr cell identification
# data = [
#     (1, "08:12:28 13 Mar 2020", "09:06:36 13 Mar 2020"),
#     (2, "09:12:48 13 Mar 2020", "09:58:16 13 Mar 2020"),
#     (3, "10:00:12 13 Mar 2020", "10:11:56 13 Mar 2020"),
#     (4, "10:14:32 13 Mar 2020", "10:29:00 13 Mar 2020"),
#     (5, "10:22:40 13 Mar 2020", "10:46:56 13 Mar 2020"),
#     (6, "10:42:32 13 Mar 2020", "10:59:04 13 Mar 2020"),
#     (7, "11:00:32 13 Mar 2020", "11:32:28 13 Mar 2020"),
#     (8, "11:30:44 13 Mar 2020", "11:51:52 13 Mar 2020"),
#     (9, "11:48:08 13 Mar 2020", "12:09:04 13 Mar 2020"),
#     (10, "12:03:08 13 Mar 2020", "12:38:04 13 Mar 2020"),
#     (11, "12:32:48 13 Mar 2020", "13:14:32 13 Mar 2020"),
#     (12, "13:53:00 13 Mar 2020", "14:05:16 13 Mar 2020"),
#     (13, "14:03:28 13 Mar 2020", "14:31:20 13 Mar 2020"),
#     (14, "15:09:36 13 Mar 2020", "15:45:20 13 Mar 2020"),
#     (15, "15:48:00 13 Mar 2020", "16:12:44 13 Mar 2020"),
#     (16, "16:12:00 13 Mar 2020", "16:33:44 13 Mar 2020"),
#     (17, "16:29:40 13 Mar 2020", "16:38:20 13 Mar 2020"),
#     (18, "16:44:24 13 Mar 2020", "17:13:32 13 Mar 2020"),
#     (19, "17:12:56 13 Mar 2020", "17:45:48 13 Mar 2020"),
#     (20, "18:10:04 13 Mar 2020", "18:28:04 13 Mar 2020"),
#     (21, "18:27:28 13 Mar 2020", "18:50:04 13 Mar 2020"),
#     (22, "19:15:32 13 Mar 2020", "19:54:00 13 Mar 2020"),
#     (23, "19:47:12 13 Mar 2020", "20:15:52 13 Mar 2020"),
#     (24, "20:14:12 13 Mar 2020", "20:48:12 13 Mar 2020"),
#     (25, "20:48:36 13 Mar 2020", "21:21:40 13 Mar 2020"),
#     (26, "21:27:40 13 Mar 2020", "21:50:48 13 Mar 2020"),
#     (27, "21:48:20 13 Mar 2020", "22:13:00 13 Mar 2020"),
#     (28, "22:28:24 13 Mar 2020", "23:00:00 13 Mar 2020"),
#     (29, "23:38:56 13 Mar 2020", "00:16:28 14 Mar 2020"),
#     (30, "00:13:24 14 Mar 2020", "00:30:44 14 Mar 2020"),
#     (31, "01:06:40 14 Mar 2020", "01:24:08 14 Mar 2020"),
#     (32, "02:19:20 14 Mar 2020", "02:34:36 14 Mar 2020"),
#     (33, "02:43:24 14 Mar 2020", "03:12:44 14 Mar 2020"),
#     (34, "03:17:16 14 Mar 2020", "03:41:20 14 Mar 2020"),
#     (35, "03:57:16 14 Mar 2020", "04:31:56 14 Mar 2020"),
#     (36, "04:27:40 14 Mar 2020", "04:50:20 14 Mar 2020"),
#     (37, "04:49:56 14 Mar 2020", "05:19:48 14 Mar 2020")
# ]


# cell-evolution based on tobac output
data = [
     (1, '07:45:00 13 Mar 2020', '08:50:00 13 Mar 2020'),
     (2, '08:35:00 13 Mar 2020', '09:50:00 13 Mar 2020'),
     (3, '09:45:00 13 Mar 2020', '11:00:00 13 Mar 2020'),
     (4, '09:35:00 13 Mar 2020', '10:35:00 13 Mar 2020'),
     (5, '10:35:00 13 Mar 2020', '10:45:00 13 Mar 2020'),
     (6, '10:45:00 13 Mar 2020', '12:30:00 13 Mar 2020'),
     (7, '11:00:00 13 Mar 2020', '12:20:00 13 Mar 2020'),
     (8, '10:20:00 13 Mar 2020', '11:55:00 13 Mar 2020'),
     (9, '11:55:00 13 Mar 2020', '12:10:00 13 Mar 2020'),
     (10, '11:10:00 13 Mar 2020', '12:50:00 13 Mar 2020'),
     (11, '11:50:00 13 Mar 2020', '12:55:00 13 Mar 2020'),
     (12, '13:05:00 13 Mar 2020', '14:20:00 13 Mar 2020'),
     (13, '14:05:00 13 Mar 2020', '14:15:00 13 Mar 2020'),
     (14, '14:45:00 13 Mar 2020', '16:05:00 13 Mar 2020'),
     (15, '15:45:00 13 Mar 2020', '16:35:00 13 Mar 2020'),
     (16, '15:20:00 13 Mar 2020', '16:25:00 13 Mar 2020'),
     (17, '16:10:00 13 Mar 2020', '18:05:00 13 Mar 2020'),
     (18, '16:40:00 13 Mar 2020', '17:00:00 13 Mar 2020'),
     (19, '16:25:00 13 Mar 2020', '18:25:00 13 Mar 2020'),
     (20, '17:25:00 13 Mar 2020', '18:20:00 13 Mar 2020'),
     (21, '15:45:00 13 Mar 2020', '18:45:00 13 Mar 2020'),
     (22, '17:35:00 13 Mar 2020', '19:55:00 13 Mar 2020'),
     (23, '18:20:00 13 Mar 2020', '19:55:00 13 Mar 2020'),
     (24, '18:45:00 13 Mar 2020', '20:45:00 13 Mar 2020'),
     (25, '20:25:00 13 Mar 2020', '21:20:00 13 Mar 2020'),
     (26, '20:45:00 13 Mar 2020', '21:55:00 13 Mar 2020'),
     (27, '21:50:00 13 Mar 2020', '22:25:00 13 Mar 2020'),
     (28, '22:25:00 13 Mar 2020', '00:00:00 14 Mar 2020'),
     (29, '23:55:00 13 Mar 2020', '00:20:00 14 Mar 2020'),
     (30, '00:00:00 14 Mar 2020', '01:05:00 14 Mar 2020'),
     (31, '00:50:00 14 Mar 2020', '01:10:00 14 Mar 2020'),
     (32, '02:20:00 14 Mar 2020', '02:35:00 14 Mar 2020'),
     (33, '02:20:00 14 Mar 2020', '03:25:00 14 Mar 2020'),
     (34, '03:20:00 14 Mar 2020', '04:20:00 14 Mar 2020'),
     (35, '04:10:00 14 Mar 2020', '05:00:00 14 Mar 2020'),
     (36, '04:10:00 14 Mar 2020', '05:00:00 14 Mar 2020'),
     (37, '04:10:00 14 Mar 2020', '05:00:00 14 Mar 2020')
]

# Iterate through each row and generate the results
for row in data:
    id = row[0]
    start_date = row[1]
    end_date = row[2]

    ct = calculate_ct(start_date, end_date)
    result = [id] + ct
    results =  ct
    print(results)
```
